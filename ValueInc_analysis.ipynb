```python
# summary of the data
data.info()
```

```python
# working with calculations
# defining variables

#costPerTransaction column calculation
#costPerTransaction = costPerItem * NumberOfItemsPurchased

#variable = dataframe['columnName]

CostPerItem = data['CostPerItem']
NumberOfItemsPurchased = data['NumberOfItemsPurchased']
CostPerTransaction = CostPerItem * NumberOfItemsPurchased
data.info()
```

```python
# Adding new column to dataframe
data['CostPerTransaction'] = CostPerTransaction

#SalesPerTransaction
data['SalesPerTransaction'] = data['SellingPricePerItem'] * data['NumberOfItemsPurchased']

#Profit = Sales - Cost
data['ProfitPerTransaction'] = data['SalesPerTransaction'] - data['CostPerTransaction']

#markup = (Sales-Cost)/Cost
data['Markup'] = (data['SalesPerTransaction'] - data['CostPerTransaction'])/data['CostPerTransaction']
```

```python
#Rounding Markup
roundMarkup = round(data['Markup'], 2)

data['Markup'] = round(data['Markup'], 2)
```

```python
#Combining data fields
data['date'] = data['Day'].astype(str)+'-'+data['Month']+'-'+data['Year'].astype(str)
data.info()
```

```python
#using split to split the clienkeyword field

split_col = data['ClientKeywords'].str.split(',',expand=True)
split_col.head(5)
```

```python
data['ClientAge'] = split_col[0]
data.shape
```

```python
data['ClientType'] = split_col[1]
data['ClientLenghtOfContract'] = split_col[2]
data.shape
```

```python
# using the replace function
data['ClientAge'] = data['ClientAge'].str.replace('[', '')
data['ClientLenghtOfContract'] = data['ClientLenghtOfContract'].str.replace(']', '')
data.head()
```

```python
data['ItemDescription'] = data['ItemDescription'].str.lower()
data.iloc[:5,7]
```

```python
# how to merge files
# bring a new data set
seasons = pd.read_csv('Data/value_inc_seasons.csv', sep=';')
```

```python
seasons.head()
```

```python
# merging files: merge_df = pd.merge(old_df, new_df, on='key')
data = pd.merge(data, seasons, on = 'Month')
data.head()
```

```python
# dropping columns
# df = df.drop('columnName', axis = 1)
data = data.drop('Year', axis = 1)
# multiple drops
data = data.drop(['ClientKeywords', 'Day', 'Month'], axis = 1)

data.info()
```

```python
# Export into a CSV
data.to_csv('ValueInc_cleaned.csv', index=False)
```
